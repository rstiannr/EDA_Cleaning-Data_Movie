# Cleaning Row

The dataset contains significant noise due to the presence of extraneous characters such as parentheses "()", pipes "|", and newline symbols "\n". These irregularities disrupt the data's consistency and structure, making it difficult to analyze or process efficiently. As a result, the dataset requires thorough cleaning and pre-processing to remove these unwanted elements and ensure that the data is formatted properly for further analysis.

```python
# clean year column from extract variable they doesn't need
df['YEAR'] = df['YEAR'].str.replace(r'[^0-9a-zA-Z]', '', regex=True)
df.info()

# check the latters
YearLatter = df[df['YEAR'].str.contains('[a-zA-Z]', regex=True, na=False)]
YearLatter
```

```python
df['YEAR_len'] = df['YEAR'].str.len()
df['YEAR_len'].value_counts()
```

```python
# Delete the latter
df['YEAR'] = df['YEAR'].str.replace(r'[^0-9]', '', regex=True)
df.info()
```

```python
# Make row with zero value be NaN
df['YEAR'] = df['YEAR'].replace('', np.nan)
df.info()
```

```python
# add . for year with range
def add_period(value):
    if isinstance(value, str) and len(value) == 8:
        return value[:4] + '.' + value[4:]
    return value

df['YEAR'] = df['YEAR'].apply(add_period)
df
```

```python
# drop column we didnt need
df = df.drop('YEAR_len', axis=1)
df.info()
```

```python
df["ONE-LINE"] = df["ONE-LINE"].str.replace('[^a-zA-Z0-9 ]','')
df["ONE-LINE"]=df["ONE-LINE"].str.strip("\n")
df["GENRE"]=df["GENRE"].str.strip("\n")

df
```

The 'Stars' column in the dataset contains both directors and actors, which are currently combined in a single field. To facilitate proper analysis, this column needs to be split into two distinct columns: one for 'Directors' and another for 'Stars' (actors). This separation will enable a more structured and accurate representation of the data, allowing for easier filtering, querying, and analysis of each role independently.


```python
df[['directors', 'stars']] = df['STARS'].str.split('Stars:', expand=True)
df.drop("STARS", inplace=True,axis=1)

df['directors'] = df['directors'].str.split(':', expand=True)[1]
df['directors'] = df['directors'].str.replace('|', '', regex=False)
df = df.replace(r'\n', '', regex=True)

df["VOTES"] = df["VOTES"].str.replace('[^0-9.]', '', regex=True)
df['VOTES'] = pd.to_numeric(df['VOTES'], errors='coerce')


df
```

# Handling Missing Value



Since the YEAR and GENRE columns contain categorical or fixed values that cannot be meaningfully replaced with the mean or any other imputation method, and the number of missing values in these columns is relatively small, we decided to drop the rows containing NaN values in these fields. This approach ensures that the integrity of the dataset is maintained, as replacing categorical data with arbitrary values would introduce inaccuracies, while removing only a small number of rows does not significantly affect the overall dataset. By dropping rows with missing values in these key columns, we maintain the quality of the data for subsequent analysis.

```python
df = df.dropna(subset=['YEAR'])
df = df.dropna(subset=['GENRE'])

df = df.drop('Gross', axis=1)
# df = df.drop('RunTime', axis=1)
```

## Observations:


**RATING**

The minimum value of 1.1 and the maximum value of 9.9 make sense within the context of movie ratings, as they fall within the expected range for such scores. The median (50th percentile) value for the RATING column is 7.1, which indicates a fairly normal distribution of ratings.

**VOTES**

**RunTime**

The minimum runtime of 1 minute and the maximum of 853 minutes seems to be reasonable. Given that 75% of the data lies below 95 minutes and the median is 60 minutes, we can expect most movies to have a runtime within a standard length. However, the high variability (standard deviation) suggests that there are some movies with significantly longer durations.

**Regarding Missing Values:**

The missing values in RATING and VOTES columns, 754 values each, can be replaced by their respective median values to ensure that the data remains consistent. The median is a robust choice for numerical columns like these, as it is less influenced by outliers compared to the mean.


```python
median_runtime = df['RunTime'].median()
df['RATING'].fillna(median_runtime, inplace=True)
df['VOTES'].fillna(median_runtime, inplace=True)
df['RunTime'].fillna(median_runtime, inplace=True)

df['directors'].fillna('None', inplace=True)
df['stars'].fillna('None', inplace=True)

df
```
